```python
import json
import time
from geopy.geocoders import Nominatim

n = 0;

with open('yearly-top100-entities-topics-sentiment1.json', 'r') as fp:
    dataset = json.load(fp)

output = open("geo.tsv","w") 
output.write("lat"+"\t"+"lon"+"\t"+"display_name"+"\t"+"freq"+"\n")



# first dictionary - years
for j in range(0, 10):
    print("year",dataset[j]["year"])

    for i in dataset[j]["places"]:

        print(dataset[j]["places"][n]["entity_name"],dataset[j]["places"][n]["freq"])

        try:
            geolocator = Nominatim()
            city = geolocator.geocode(dataset[j]["places"][n]["entity_name"], language='en')

            print(city.raw["lat"],city.raw["lon"],city.raw["display_name"])
            output.write(city.raw["lat"]+"\t"+city.raw["lon"]+"\t"+city.raw["display_name"]+"\t"+ str(dataset[j]["places"][n]["freq"])+"\n")

            time.sleep(0.3)
        except Exception as e:
            print("There's an error! ", e )

        n +=1

output.close()
    
```
